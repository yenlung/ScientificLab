# 16. 用數學的方式算數學

我們來做所謂的「符號式」運算。

## 符號式計算套件 SymPy

````python
from sympy import *
````

## 讓式子美化的魔術

````python
init_printing()
````

## 要先設什麼是變數

我們要告訴 Python, 例如 x 是一個變數。

````python
x = symbols("x")
````

## 求反導數

````python
integrate(1/(x**2-1))
````

## 定積分

````python
integrate(1/(x**2-1), (x, 2,3))
````

## 秀式子

````python
Integral(1/(x**2-1), (x,2,3))
````

## 互動一下

````python
def ntaylor(n):
    expr = sin(x).series(x,0,n).removeO()
    display(expr)
    f = lambdify(x, expr, "numpy")
    a = linspace(-10,10,300)
    plt.plot(a, f(a));
    plt.ylim(-1,1);
````

## 開始!

````python
interact(ntaylor, n = (1,30));
````
